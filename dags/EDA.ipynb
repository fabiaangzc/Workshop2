```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_path = "data/spotify_dataset.csv"
grammys_path = "data/the_grammy_awards.csv"

spotify = pd.read_csv(spotify_path)
grammys = pd.read_csv(grammys_path)
```

# EDA 


```python
# Dimensiones y tipos de datos
print('Spotify shape:', spotify.shape)
print('Grammys shape:', grammys.shape)

print('\nSpotify info:')
spotify.info()
print('\nGrammys info:')
grammys.info()

print('\nPrimeras filas Spotify:')
display(spotify.head())
print('\nPrimeras filas Grammys:')
display(grammys.head())
```

```python
print('\nDuplicados Spotify:', spotify.duplicated().sum())
print('Duplicados Grammys:', grammys.duplicated().sum())
```

```python
# Conteo de nulos y duplicados
print('Nulos Spotify:')
print(spotify.isnull().sum().sort_values(ascending=False))
print('\nNulos Grammys:')
print(grammys.isnull().sum().sort_values(ascending=False))
```

```python
# Ejemplos de artistas y frecuencia
print('Ejemplos de artistas Spotify:')
print(spotify['artists'].dropna().unique()[:10])
print('\nEjemplos de artistas Grammys:')
print(grammys['artist'].dropna().unique()[:10])

# Frecuencia de aparición de artistas
print('\nTop artistas Spotify:')
print(spotify['artists'].value_counts().head(10))
print('\nTop artistas Grammys:')
print(grammys['artist'].value_counts().head(10))

# Porcentaje de filas con múltiples artistas
multi_spotify = spotify['artists'].fillna('').str.contains(';').mean()
print(f'Porcentaje de múltiples artistas en Spotify: {multi_spotify:.2%}')
```

```python
# Spotify: distribución de géneros
if 'track_genre' in spotify.columns:
    top_genres = spotify['track_genre'].value_counts().head(15)
    sns.barplot(y=top_genres.index, x=top_genres.values)
    plt.title('Top 15 géneros Spotify')
    plt.show()

# Grammys: distribución de categorías
top_cats = grammys['category'].value_counts().head(15)
sns.barplot(y=top_cats.index, x=top_cats.values)
plt.title('Top 15 categorías Grammys')
plt.show()

# Grammys: conteo de winners
if 'winner' in grammys.columns:
    print('Conteo de winners:')
    print(grammys['winner'].value_counts())
```

```python
# Normalización básica de artistas para comparar
import re, unicodedata
def normalize_artist(name):
    if pd.isna(name) or str(name).strip() == "":
        return ""
    s = str(name).lower()
    s = ''.join(ch for ch in unicodedata.normalize('NFKD', s) if not unicodedata.combining(ch))
    s = re.sub(r"\(.*?\)|\[.*?\]", "", s)
    s = re.split(r"\s+(?:feat\.|featuring|ft\.|feat|featuring:|feats)\s+", s)[0]
    s = s.replace('&', ' and ')
    s = re.sub(r"[^a-z0-9\s]", '', s)
    s = re.sub(r"^the\s+", '', s)
    s = re.sub(r"\s+", ' ', s).strip()
    return s

spotify_primary = spotify['artists'].dropna().apply(lambda x: normalize_artist(x.split(';')[0]))
grammys_artists = grammys['artist'].dropna().apply(normalize_artist)

spotify_set = set([a for a in spotify_primary if a])
grammys_set = set([a for a in grammys_artists if a])

intersection = spotify_set & grammys_set
union = spotify_set | grammys_set

print(f"Artistas únicos Spotify: {len(spotify_set)}")
print(f"Artistas únicos Grammys: {len(grammys_set)}")
print(f"Artistas en común: {len(intersection)}")
print(f"Coverage (intersection/grammys): {len(intersection)/len(grammys_set):.2%}")
print(f"Coverage (intersection/spotify): {len(intersection)/len(spotify_set):.2%}")
print(f"Jaccard similarity: {len(intersection)/len(union):.2%}")

# Ejemplos de artistas no coincidentes
grammys_only = list(grammys_set - spotify_set)[:10]
spotify_only = list(spotify_set - grammys_set)[:10]
print('\nEjemplos de artistas Grammys no encontrados en Spotify:', grammys_only)
print('Ejemplos de artistas Spotify no encontrados en Grammys:', spotify_only)
```

# Conclusiones y decisiones — Resumen del EDA

## 1) Calidad general
- Spotify: muy pocos nulos (3), 0 duplicados, ~114k filas. Dataset listo para analizar, eliminar `Unnamed: 0` y conservar columnas relevantes.  
- Grammys: ~4.8k filas; nulos relevantes en `artist` (38%), `workers` (45%), `img` (28%); 0 duplicados. `winner` siempre True → solo ganadores (importantísimo para la interpretación de métricas).

## 2) Artistas y joinability
- Spotify ~17.4k artistas únicos; Grammys ~1.56k; intersección = 477.  
- Coverage Grammys ≈ 30.5% → casi 1/3 de los artistas Grammy aparecen en Spotify. Coverage Spotify ≈ 2.7%. Jaccard ≈ 2.6%: conjuntos bastante distintos.
- Hay muchos registros con múltiples artistas en Spotify (~26%) y casos compuestos en Grammys. Necesario "explosionar" artistas y normalizar.

## 3) Variables de negocio — qué conservar
- Conservar en Spotify: `track_id`, `track_name`, `artists`, `album_name`, `popularity`, `duration_ms`, `track_genre` (y opcionalmente features de audio si se van a usar).  
- Conservar en Grammys: `year`, `category`, `artist`, `winner`.  
- Descartar/archivar: `img`, `workers`, `title`, `published_at`, `updated_at` (no útiles para DW en este alcance).

## 4) Reglas y decisiones para la transformación (implementación)
- Normalización de artistas: lower, quitar tildes, remover paréntesis/brackets, cortar sufijos `feat.`/`ft.`, reemplazar `&` por `and`.  
- Explosión de artistas: separar por `;` (y otras conjunciones) para tener una fila por artista en las tablas de dimensión/fact (o una tabla puente).  
- Crear `artist_key` normalizado y `dim_artist` (id, artist_key, display_name, aliases).  
- Crear `dim_award` (award_id, year, category, category_norm).  
- Crear `fact_streaming` (por track/por artista) y `fact_awards` (por award win).  

## 5) Reglas de negocio sobre missing/join
- Si un artista Grammy no aparece en Spotify → incluirlo en `dim_artist` con métricas Spotify nulas.  
- Si un artista Spotify no aparece en Grammys → mantenerlo en `dim_artist` con `awards_count = 0`.  
- Al unir, operar a nivel `artist_key` (no por cadena sin normalizar); documentar emparejamientos manuales o fuzzy que aceptes.

## 6) KPIs y uso futuro
- KPIs directos: porcentaje de ganadores presentes en Spotify; popularidad media en Spotify de ganadores; distribución por categoría/año.  
- Nota: porque Grammys contiene solo ganadores, las comparaciones deben interpretarse como "éxito de ganadores en Spotify" y no como comparación ganadores vs nominados.
